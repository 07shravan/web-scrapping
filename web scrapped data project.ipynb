```python
import os
import time
import requests
import sys

def html_data():
    for year in range(2013,2019):
        for month in range(1,13):
            if month <10:
                url="https://en.tutiempo.net/climate/0{}-{}/ws-432950.html".format(month,year)
                
            else:
                url="https://en.tutiempo.net/climate/{}-{}/ws-432950.html".format(month,year)
               
            texts= requests.get(url) 
            text_utf=texts.text.encode("utf=8")
            if not os.path.exists("C:\Data\html\{}".format(year)):
                os.makedirs("C:\Data\html\{}".format(year))
            with open("C:\Data\html\{}\{}.html".format(year,month),"wb") as output:
                output.write(text_utf)
            sys.stdout.flush()
   
        
            
if __name__ =="__main__":
    starttime= time.time()
    html_data()
    endtime= time.time()
    print("total time taken_{}".format(endtime-starttime))
                
        
        
        
        
        
        
```

```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup as bs

avg=[]
final=[]


for year in {2013}:
    for month in range(1,13):
        file_html =open("C:\Data\html\{}\{}.html".format(year,month),"rb")
        plain_text=file_html.read()
        temp=[]
        soup=bs(plain_text,"lxml")
        
        for table in soup.find_all("table",{"class":"medias mensuales numspan"}):
            for tbody in table:
                for tr in tbody:
                    a=tr.get_text()
                    temp.append(a)
    
        rows=len(temp)/15
        for times in range(round(rows)):
            newtemp=[]
            for i in range(15):
                newtemp.append(temp[0])
                temp.pop(0)
            final.append(newtemp)
    
        
    data=pd.read_csv(r"C:\AQI\aqi{}.csv".format(year))
    data["PM2.5"]=np.where(data["PM2.5"].str.isalpha(),0,data["PM2.5"])
    data["PM2.5"]=np.where(data["PM2.5"].isin(["---"]),0,data["PM2.5"])
    data["PM2.5"]=np.round(data["PM2.5"].astype(np.float32),2)
    data["Date"]=[val[0] for val in data["Date"].str.split(" ",1)]
    data["Date"]=pd.to_datetime(data["Date"]).dt.date
    avg.append(data.groupby("Date")["PM2.5"].mean().reset_index()["PM2.5"])

final=pd.DataFrame(final)
final=final.rename(columns=final.iloc[0,:])
final=final.drop( final[final["Day"]=="Monthly means and totals:"].index)
final=final.drop( final[final["Day"]=="Day"].index).reset_index()
avg=pd.DataFrame(avg).transpose()
final2013=pd.concat([final,avg],axis=1)
final2013=final2013.drop(columns=["SLP","index","VG","RA","SN","TS","FG"],axis=1)
final2013.to_csv("real_csv1")


   

    
        
```
